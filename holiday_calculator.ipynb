# Holiday Date Calculator

This notebook computes the dates of various holidays for a range of years. It includes both fixed-date holidays and those that are calculated based on lunar cycles, Easter, or other methods.

```python
import calendar
from datetime import date, timedelta

def get_easter_date(year):
    """Calculate Easter date for a given year using the Anonymous Gregorian algorithm."""
    a = year % 19
    b = year // 100
    c = year % 100
    d = b // 4
    e = b % 4
    f = (b + 8) // 25
    g = (b - f + 1) // 3
    h = (19 * a + b - d - g + 15) % 30
    i = c // 4
    k = c % 4
    l = (32 + 2 * e + 2 * i - h - k) % 7
    m = (a + 11 * h + 22 * l) // 451
    month = (h + l - 7 * m + 114) // 31
    day = ((h + l - 7 * m + 114) % 31) + 1
    return date(year, month, day)

def get_mlk_day(year):
    """Martin Luther King Jr. Day: Third Monday in January."""
    first_day = date(year, 1, 1)
    days_until_monday = (0 - first_day.weekday()) % 7
    first_monday = first_day + timedelta(days=days_until_monday)
    return first_monday + timedelta(weeks=2)

def get_presidents_day(year):
    """Presidents' Day: Third Monday in February."""
    first_day = date(year, 2, 1)
    days_until_monday = (0 - first_day.weekday()) % 7
    first_monday = first_day + timedelta(days=days_until_monday)
    return first_monday + timedelta(weeks=2)

def get_memorial_day(year):
    """Memorial Day: Last Monday in May."""
    last_day = date(year, 6, 1) - timedelta(days=1)
    days_since_monday = (last_day.weekday() - 0) % 7
    return last_day - timedelta(days=days_since_monday)

def get_labor_day(year):
    """Labor Day: First Monday in September."""
    first_day = date(year, 9, 1)
    days_until_monday = (0 - first_day.weekday()) % 7
    return first_day + timedelta(days=days_until_monday)

def get_columbus_day(year):
    """Columbus Day: Second Monday in October."""
    first_day = date(year, 10, 1)
    days_until_monday = (0 - first_day.weekday()) % 7
    return first_day + timedelta(days=days_until_monday) + timedelta(weeks=1)

def get_thanksgiving(year):
    """Thanksgiving: Fourth Thursday in November."""
    first_day = date(year, 11, 1)
    days_until_thursday = (3 - first_day.weekday()) % 7
    first_thursday = first_day + timedelta(days=days_until_thursday)
    return first_thursday + timedelta(weeks=3)
```

```python
# Define holidays with their calculation functions
HOLIDAYS = {
    "New Year's Day": lambda y: date(y, 1, 1),
    "Martin Luther King Jr. Day": get_mlk_day,
    "Presidents' Day": get_presidents_day,
    "Valentine's Day": lambda y: date(y, 2, 14),
    "St. Patrick's Day": lambda y: date(y, 3, 17),
    "Easter": get_easter_date,
    "Memorial Day": get_memorial_day,
    "Independence Day": lambda y: date(y, 7, 4),
    "Labor Day": get_labor_day,
    "Columbus Day": get_columbus_day,
    "Veterans Day": lambda y: date(y, 11, 11),
    "Thanksgiving": get_thanksgiving,
    "Christmas Day": lambda y: date(y, 12, 25),
    "New Year's Eve": lambda y: date(y, 12, 31),
}
```

```python
# Function to compute holiday table for a range of years
def compute_holiday_table(start_year, end_year):
    """Compute a table of holiday dates for the given year range."""
    data = []
    
    for year in range(start_year, end_year + 1):
        row = {'Year': year}
        for holiday_name, calc_func in HOLIDAYS.items():
            try:
                holiday_date = calc_func(year)
                row[holiday_name] = holiday_date.strftime('%Y-%m-%d (%A)')
            except Exception as e:
                row[holiday_name] = f'Error: {e}'
        data.append(row)
    
    return data
```

```python
# Example: Compute holidays for 2024-2028
holiday_data = compute_holiday_table(2024, 2028)
```

```python
# Display the table
import pandas as pd

df = pd.DataFrame(holiday_data)
# Reorder columns to have Year first
cols = ['Year'] + [c for c in df.columns if c != 'Year']
df = df[cols]

# Style only holiday date columns (skip Year)
holiday_columns = df.columns[1:]

def highlight_weekend(val):
    if 'Saturday' in str(val) or 'Sunday' in str(val):
        return 'background-color: #ffcccb'
    return ''

styled_df = df.style.map(highlight_weekend, subset=pd.IndexSlice[:, holiday_columns])

styled_df
```

```python
# Display holidays for a specific year
def display_year_holidays(year):
    """Display holidays for a specific year in a nicely formatted way."""
    print(f"\n{'='*60}")
    print(f"Holidays in {year}")
    print('='*60)
    
    for holiday_name, calc_func in sorted(HOLIDAYS.items(), key=lambda x: (x[1](year).month, x[1](year).day)):
        try:
            holiday_date = calc_func(year)
            day_name = calendar.day_name[holiday_date.weekday()]
            print(f"{holiday_name:25s}: {holiday_date.strftime('%B %d, %Y')} ({day_name})")
        except Exception as e:
            print(f"{holiday_name:25s}: Error - {e}")

# Example: Display 2024 holidays
display_year_holidays(2024)
```
